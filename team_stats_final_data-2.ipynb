```python
df_wins = pd.read_csv("/content/team_wins_by_year.csv")
df_wins.head()
```

```python
df_wins = df_wins.drop(['Unnamed: 0'],axis=1)
```

```python
df_wins.rename(columns = {'Team Name':'Team Name1'}, inplace = True)
```

```python
df_wins.head()
```

```python
df_wins = df_wins.sort_values(['Year','Team Name1'],ascending = [True,True])
```

```python
df = df.sort_values(['Year','Team Name'],ascending = [True,True])
```

```python
df.reset_index(inplace=True)
df_wins.reset_index(inplace=True)

```

```python
df = df.drop(['index'],axis=1)
df_wins = df_wins.drop(['index'],axis=1)
```

```python
df.head(10)
```

```python
df_wins.head(10)
```

```python
result = pd.concat([df, df_wins], axis=1, join='inner')
display(result)
```

```python
result = result.iloc[:,:-1]
```

```python
result = result.drop(['Team Name1'], axis=1)
```

```python
winPercentage= []

for i in range(len(result)):
    winPercent = (result.loc[i,"Wins"]/result.loc[i,"Games Played"])*100
    winPercentage.append(winPercent)
    
result['Win Percentage'] = winPercentage
```

```python
lossPercentage= []

for i in range(len(result)):
    lossPercent = (100 - result.loc[i,"Win Percentage"])
    lossPercentage.append(lossPercent)
    
result['Loss Percentage'] = lossPercentage
```

```python
result = result.round({"Loss Percentage":2, "Win Percentage":2})
```

```python
result = result.reindex(columns=['Team Name', 'Year', 'Games Played',
                                 'Minutes','Wins','Loss',	'Win Percentage',	'Loss Percentage','Points', 
                                 'Three Pointers', 'Two Pointers',
                                 'Second Chance Points','Counter Attack Points', 'Free Throws',	'Blocks',	'Blocked Attempts',
                                 'Personal Fouls','Flagrant Fouls', 'Team Technical Fouls',
                                 'Team Fouls','Total Fouls','Total Rebounds','Rebounds', 'Assists',	'Turnovers',	
                                 'Assists Turnover Ratio','Steals'])
```
