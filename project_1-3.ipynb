```python
def teams_data(yearID,keyID,teamID):

  series_stats1 = requests.get("https://api.sportradar.us/nba/trial/v7/en/seasons/"+yearID+"/REG/teams/"+teamID+"/statistics.json?api_key="+keyID)
  series_stats1 = series_stats1.json()
  team_name = series_stats1['market'] +" "+ series_stats1['name']
  total1=series_stats1['own_record']['total']

  dic= {'Team Name' : team_name,'Year': yearID,'Games Played':total1['games_played'], 'Minutes': total1['minutes']
                  ,'Free Throws' : total1['free_throws_made']
                  ,'Points' : total1['points'], 'Three Pointers' : total1['three_points_made'],'Second Chance Points' : total1['second_chance_pts']
                  ,'Counter Attack Points' : total1['fast_break_pts'],'Blocks' : total1['blocks'], 'Blocked Attempts': total1['blocked_att']
                  ,'Personal Fouls' : total1['personal_fouls'], 'Flagrant Fouls': total1['flagrant_fouls']
                  ,'Team Technical Fouls' : total1['team_tech_fouls'],'Team Fouls': total1['team_fouls']
                  ,'Total Rebounds' : total1['total_rebounds'],'Total Fouls' : total1['total_fouls']
                  ,'Assists' : total1['assists'],'Turnovers' : total1['turnovers'],'Assists Turnover Ratio' : total1['assists_turnover_ratio'],'Steals' : total1['steals'], 'Two Pointers': total1['two_points_made']
                  ,'Rebounds': total1['rebounds']
           }

  return dic
```
